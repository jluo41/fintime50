# Actual Operation

```python
# we need headers to disguise our bot as a browser

headers = {
    "Connection": "keep-alive",
    "Cache-Control": "max-age=0",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "User-Agent": "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/34.0.1847.131 Safari/537.36",
    "Accept-Encoding": "gzip,deflate,sdch",
    "Accept-Language": "zh-CN,zh;q=0.8,en-US;q=0.6,en;q=0.4,zh-TW;q=0.2",
}
```

```python
import requests
from scrapy.http import TextResponse
```

```python
# add header here, therefore you can get the right url
# notice that I define the headers dictionary in In[1] unit.
r = requests.get('http://www.journals.elsevier.com/decision-support-systems/', 
                 headers = headers)

response = TextResponse(r.url, body = r.text, encoding = 'utf-8')
response
```

the type of the 'response' as above is 200, which means that the url is valid

```python
response.body
```

**Copy Xpath**

![](getXpath.png)

```python
# In this way, we can get the xpath for that element
XpathForAuthor = '//*[@id="Title"]/div[2]/div/div[4]/span[2]'
```

```python
# there are the xpathes.

issn_xpath = '//*[@class="issn keyword"]/span/text()'
chief_editor_xpath = '//*[@id="Title"]//span[@class="nowrap"]/text()'
title_xpath = '//*[@id="Title"]//h1[@itemprop="name"]/text()'
description_xpath = '//*[@class="publication-description"]//p'
coverimage_xpath = '//*[@id="Title"]//img[@class="cover-img"]/@src'
```

```python
# return a list, and the type of elements is Selector.

response.xpath(issn_xpath)
```

```python
# in this way extract the values from Selectors.
# return a list here

response.xpath(issn_xpath).extract()
```
